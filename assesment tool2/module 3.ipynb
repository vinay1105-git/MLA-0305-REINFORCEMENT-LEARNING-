```python
import pandas as pd

# Rewards
rewards = {
    ("S1", "A1", "S2"): 5,
    ("S1", "A2", "S2"): 10,
    ("S1", "A1", "S3"): -1,
    ("S1", "A2", "S3"): -5,

    ("S2", "A1", "S1"): 3,
    ("S2", "A2", "S1"): 7,
    ("S2", "A1", "S3"): 2,
    ("S2", "A2", "S3"): 1,

    ("S3", "A1", "S1"): 4,
    ("S3", "A2", "S1"): 6,
    ("S3", "A1", "S2"): 0,
    ("S3", "A2", "S2"): -2
}

# Create reward matrix table
reward_table = pd.DataFrame(
    [
        [current_state, action, next_state, reward]
        for (current_state, action, next_state), reward in rewards.items()
    ],
    columns=["Current State", "Action", "Next State", "Reward"]
)

print("=" * 60)
print("REWARD MATRIX")
print("=" * 60)

display(reward_table)
```
