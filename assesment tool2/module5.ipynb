```python
import pandas as pd

states = ["S1", "S2", "S3"]
actions = ["A1", "A2"]

transition_probabilities = {
    ("S1","A1","S2"):0.6, ("S1","A1","S3"):0.2,
    ("S1","A2","S2"):0.4, ("S1","A2","S3"):0.8,

    ("S2","A1","S1"):0.7, ("S2","A1","S3"):0.5,
    ("S2","A2","S1"):0.3, ("S2","A2","S3"):0.5,

    ("S3","A1","S1"):0.9, ("S3","A1","S2"):0.4,
    ("S3","A2","S1"):0.1, ("S3","A2","S2"):0.6
}

rewards = {
    ("S1","A1","S2"):5, ("S1","A1","S3"):-1,
    ("S1","A2","S2"):10, ("S1","A2","S3"):-5,

    ("S2","A1","S1"):3, ("S2","A1","S3"):2,
    ("S2","A2","S1"):7, ("S2","A2","S3"):1,

    ("S3","A1","S1"):4, ("S3","A1","S2"):0,
    ("S3","A2","S1"):6, ("S3","A2","S2"):-2
}

summary = []

for state in states:
    for action in actions:
        expected_reward = 0

        for (s, a, next_state), prob in transition_probabilities.items():
            if s == state and a == action:
                expected_reward += prob * rewards[(s, a, next_state)]

        summary.append([state, action, round(expected_reward, 2)])

summary_df = pd.DataFrame(
    summary,
    columns=["Current State", "Action", "Expected Immediate Reward"]
)

print("=" * 65)
print("EXPECTED IMMEDIATE REWARD SUMMARY")
print("=" * 65)

display(summary_df)
```
