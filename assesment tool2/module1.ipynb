```python
# ==========================================
# Markov Decision Process (MDP)
# ==========================================

# States
states = ["S1", "S2", "S3"]

# Actions
actions = ["A1", "A2"]

# Transition Probabilities
transition_probabilities = {
    ("S1", "A1", "S2"): 0.6,
    ("S1", "A1", "S3"): 0.2,
    ("S1", "A2", "S2"): 0.4,
    ("S1", "A2", "S3"): 0.8,

    ("S2", "A1", "S1"): 0.7,
    ("S2", "A1", "S3"): 0.5,
    ("S2", "A2", "S1"): 0.3,
    ("S2", "A2", "S3"): 0.5,

    ("S3", "A1", "S1"): 0.9,
    ("S3", "A1", "S2"): 0.4,
    ("S3", "A2", "S1"): 0.1,
    ("S3", "A2", "S2"): 0.6
}

# Rewards
rewards = {
    ("S1", "A1", "S2"): 5,
    ("S1", "A2", "S2"): 10,
    ("S1", "A1", "S3"): -1,
    ("S1", "A2", "S3"): -5,

    ("S2", "A1", "S1"): 3,
    ("S2", "A2", "S1"): 7,
    ("S2", "A1", "S3"): 2,
    ("S2", "A2", "S3"): 1,

    ("S3", "A1", "S1"): 4,
    ("S3", "A2", "S1"): 6,
    ("S3", "A1", "S2"): 0,
    ("S3", "A2", "S2"): -2
}

# ==========================================
# Display Information
# ==========================================

print("="*65)
print("      MARKOV DECISION PROCESS (MDP) SUMMARY")
print("="*65)

print("\nLEGEND")
print("-"*65)
print("S1, S2, S3  : States")
print("A1, A2      : Actions")
print("P(s'|s,a)   : Transition Probability")
print("R(s,a,s')   : Reward received after taking action")
print("s           : Current State")
print("a           : Action")
print("s'          : Next State")

print("\nMDP DETAILS")
print("-"*65)
print("Number of States                 :", len(states))
print("Number of Actions                :", len(actions))
print("Number of Transition Probabilities:", len(transition_probabilities))
print("Number of Rewards                :", len(rewards))

print("\nStates :", states)
print("Actions:", actions)

print("\nTRANSITION PROBABILITIES")
print("-"*65)
for key, value in transition_probabilities.items():
    s, a, ns = key
    print(f"P({ns} | {s}, {a}) = {value}")

print("\nREWARDS")
print("-"*65)
for key, value in rewards.items():
    s, a, ns = key
    print(f"R({s}, {a}, {ns}) = {value}")

print("\n" + "="*65)
print("SUMMARY")
print("="*65)
print(f"Total States                  : {len(states)}")
print(f"Total Actions                 : {len(actions)}")
print(f"Transition Probability Entries: {len(transition_probabilities)}")
print(f"Reward Entries                : {len(rewards)}")
```
