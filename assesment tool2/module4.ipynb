```python
# Rewards
rewards = {
    ("S1", "A1", "S2"): 5,
    ("S1", "A2", "S2"): 10,
    ("S1", "A1", "S3"): -1,
    ("S1", "A2", "S3"): -5,

    ("S2", "A1", "S1"): 3,
    ("S2", "A2", "S1"): 7,
    ("S2", "A1", "S3"): 2,
    ("S2", "A2", "S3"): 1,

    ("S3", "A1", "S1"): 4,
    ("S3", "A2", "S1"): 6,
    ("S3", "A1", "S2"): 0,
    ("S3", "A2", "S2"): -2
}

# Transition Probabilities
transition_probabilities = {
    ("S1", "A1", "S2"): 0.6,
    ("S1", "A1", "S3"): 0.2,
    ("S1", "A2", "S2"): 0.4,
    ("S1", "A2", "S3"): 0.8,

    ("S2", "A1", "S1"): 0.7,
    ("S2", "A1", "S3"): 0.5,
    ("S2", "A2", "S1"): 0.3,
    ("S2", "A2", "S3"): 0.5,

    ("S3", "A1", "S1"): 0.9,
    ("S3", "A1", "S2"): 0.4,
    ("S3", "A2", "S1"): 0.1,
    ("S3", "A2", "S2"): 0.6
}

states = ["S1", "S2", "S3"]
actions = ["A1", "A2"]

print("=" * 75)
print("EXPECTED IMMEDIATE REWARD CALCULATION")
print("=" * 75)

for state in states:
    print(f"\nCurrent State : {state}")
    print(f"Number of Actions : {len(actions)}")

    for action in actions:

        expected_reward = 0
        calculation = []

        for (s, a, next_state), probability in transition_probabilities.items():
            if s == state and a == action:
                reward = rewards[(s, a, next_state)]
                expected_reward += probability * reward
                calculation.append(f"({probability} × {reward})")

        print(f"\nAction : {action}")
        print("Calculation :", " + ".join(calculation))
        print(f"Expected Immediate Reward = {expected_reward:.2f}")

print("\n" + "=" * 75)
print("Calculation Completed Successfully")
```
